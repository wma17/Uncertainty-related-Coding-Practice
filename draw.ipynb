```python
import matplotlib.pyplot as plt

# Data
samples = ['G.png', 'E.png', 'J.png', 'D.png', '1F.png', '6I.png', '2H.png', 'C.png', '3B.png', 'A.png']
single_model_probs = [0.9749, 1.0000, 0.9964, 0.7699, 0.6075, 0.7395, 0.3729, 0.6074, 0.9609, 0.7945]
ensemble_model_probs = [0.7833, 0.4629, 0.9670, 0.6162, 0.7789, 0.6868, 0.7240, 0.4749, 0.3802, 0.3116]

# X-axis positions
x_positions = range(len(samples))

# Create the plot
plt.figure(figsize=(10, 6))

# Single Model Scatter Plot
plt.scatter(x_positions, single_model_probs, color='blue', marker='x', label='Single Model')

# Ensemble Model Scatter Plot
plt.scatter(x_positions, ensemble_model_probs, color='orange', marker='x', label='MC-Dropout Model')

# Customize the plot
plt.xlabel('Sample Name')
plt.ylabel('Prediction Probability')
plt.title('Prediction Probabilities: Single Model vs MC-Dropout Model')
plt.xticks(ticks=x_positions, labels=samples, rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()

# Display the plot
plt.show()
```
